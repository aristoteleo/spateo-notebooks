```python
import sys
sys.path.append("/jdfssz1/ST_SUPERCELLS/P20Z10200N0059/panhailin/pycharm_bin/icell_qiu/") # the dir of sequencing fold
from sequencing.icell import Icell
import cv2
import numpy as np
```

```python
infile = "SS200000135TL_D1.txt.sorted"
ins = Icell(infile)
cellsize = 25 # expected cell diameter in pixels
ins.readData()
```

```python
# way 1 and way2
ins.saveRawAstif(f"all.intron.xrange_{ins.xmin}-{ins.xmax}_yrange_{ins.ymin}-{ins.ymax}.tif")
ins.gBlurAndSaveToTif(ks=cellsize)
cutoff = ins.getKneeOfgBlur() # way 1
#cutoff = 25 #way2
ins.maskCellFromgBlur(cutoff=cutoff)
ins.mclose(kernelSize=int(cellsize/4), outFig="cellMaskc.tif")
ins.mopen(kernelSize=int(cellsize/4), outFig="cellMaskco.tif")
ins.getCellLabels(kernelgblur=cellsize, min_distance=int(cellsize*0.8))
ins.enlarge(kpixel=7,maxA=625)
ins.saveCellLabels()
ins.addCellLabelsToMatrix("expressionCellLabels.matrix")
ins.assignNonCellSpots(radius=11)
```

```python
# way3
ins.saveRawAstif(f"all.intron.xrange_{ins.xmin}-{ins.xmax}_yrange_{ins.ymin}-{ins.ymax}.tif")
tissueM = cv2.imread("tissueMask.tif", 0) # 0: cell   255: background
#ins.rawdata[tissueM==255] = 0
ins.nbnEM(k=3, w=np.array([0.99,0.01]), mu=np.array([10.0,2000.0]), var=np.array([20.0,4000.0]), tissueM=tissueM)
ins.gBlurAndSaveToTif(ks=cellsize, usenbnEM=True)
cutoff = ins.getKneeOfgBlur()
ins.maskCellFromgBlur(cutoff=cutoff)
ins.mclose(kernelSize=int(cellsize/4), outFig="cellMaskc.tif")
ins.mopen(kernelSize=int(cellsize/4), outFig="cellMaskco.tif")
ins.getCellLabels(kernelgblur=cellsize, min_distance=int(cellsize*0.8))
ins.enlarge(kpixel=7,maxA=625)
ins.saveCellLabels()
ins.addCellLabelsToMatrix("expressionCellLabels.matrix")
ins.assignNonCellSpots(radius=11)
```

```python
#way4
tissueM = cv2.imread("tissueMask.tif", 0)
ins.rawdata[tissueM==255] = 0
ins.nbnEM(k=3, w=np.array([0.99,0.01]), mu=np.array([10.0,2000.0]), var=np.array([20.0,4000.0]))
ins.maskCellFromEM(cutoff=230)
ins.mclose(kernelSize=int(cellsize/4), outFig="cellMaskc.tif")
ins.mopen(kernelSize=int(cellsize/4), outFig="cellMaskco.tif")
ins.getCellLabels(kernelgblur=cellsize, min_distance=int(cellsize*0.8))
ins.enlarge(kpixel=7,maxA=625)
ins.saveCellLabels()
ins.addCellLabelsToMatrix("expressionCellLabels.matrix")
ins.assignNonCellSpots(radius=11)
```
