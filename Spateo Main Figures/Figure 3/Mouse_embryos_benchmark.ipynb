```python
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib import style
import pandas as pd
import numpy as np
```

```python
res_df = pd.read_pickle('Mouse_benchmark_results.pkl')
res_df
```

```python
my_palette = [np.array([157,60,135]) / 255] + sns.color_palette('Paired')[0:3]+sns.color_palette('Paired')[4:5]
```

```python
import matplotlib as mpl
sns.set_theme(style="ticks")
COLOR = 'black'
mpl.rcParams['text.color'] = COLOR
mpl.rcParams['axes.labelcolor'] = COLOR
mpl.rcParams['xtick.color'] = COLOR
mpl.rcParams['ytick.color'] = COLOR
mpl.rcParams.update({'text.color' : COLOR,
                     'axes.labelcolor' : COLOR})
fig, ax = plt.subplots(1, 1,figsize=(3.2,2))
axs = [ax]
g = sns.lineplot(
    x="Proportion", 
    y="Accuracy",
    hue="Method",
    data=res_df[res_df['Time']  == 95],
    ax=axs[0],
    err_kws={'edgecolor': None},
    palette=my_palette,
    legend=False,
    
)
# axs[0].legend(ncol=3, loc='upper center', 
#                 bbox_to_anchor=(0.5, 1.3), frameon=False, columnspacing=0.5,numpoints=10)
axs[0].spines['bottom'].set_color(COLOR)
axs[0].spines['top'].set_color(COLOR) 
axs[0].spines['right'].set_color(COLOR)
axs[0].spines['left'].set_color(COLOR)
sns.despine(top=True, right=True)
axs[0].set_xlabel('Cumulative distribution', color='black')
axs[0].set_ylabel('CLC score', color='black')
# axs[0].set_yscale("log")

# plt.xscale("log")
import os
plt.savefig(os.path.join('D:/00ScienceStudy/Biology/Spateo/Figure_design/Fig3_new/figures', 'mouse_benchmark_new_95.pdf'), dpi=300, bbox_inches='tight', transparent=True)
```
