```python
method_name = 'SLAT_GT'
# ori_results_folder = ""
data_folder = f"./global_results/{method_name}/"
results_folder = f"./global_results/{method_name}_global/"
if not os.path.exists(results_folder):
    os.mkdir(results_folder)
```

```python
for i in range(16):
    # i = 1
    # data_name = "sagittal{}".format(i)
    data_name = "well_{:02d}".format(i)
    cur_data_folder = os.path.join(data_folder, data_name)
    s = ad.read_h5ad(os.path.join(data_folder, data_name + '.h5ad'))
    slices = [s[s.obs['slice_id'] == str(i)] for i in range(10)]
    key_added = "global_align_spatial"
    global_align_slices = st.align.morpho_global_optimization(
        models=slices,
        spatial_key='align_spatial',
        key_added="global_align_spatial",
        device="3",
        mode="SN-S",
        max_iter_global=5,
        max_iter=200,
        verbose=False,
        partial_robust_level=50,
        beta=1e-4,
        lambdaVF=1e4,
        vis_optimization=False,
        fps=10,
        stepsize=4,
        nn_init=True,
        batch_size=4000,
        nonrigid_idx=None,
        neighbor_size=2,
    )
        
    stacked_slices = ad.concat(global_align_slices, axis=0, label='slice_id')
    stacked_slices.write_h5ad(os.path.join(results_folder, data_name+'.h5ad'))
    
```
