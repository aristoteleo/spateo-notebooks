```python
import matplotlib.colors as clr
import matplotlib.pyplot as plt
import scanpy as sc
adata = sc.read('/zfs-1/zuolulu/spatialDEG/test/human_cortex/results.h5ad')
color_self = clr.LinearSegmentedColormap.from_list('', ["grey",'pink', "red"], N=256)
sc.pl.scatter(adata,alpha=1,x="y_array",y="x_array",color="MBP",title='MBP',color_map=color_self,show=True,size=300000/adata.shape[0])
sc.pl.scatter(adata,alpha=1,x="y_array",y="x_array",color="CNP",title='CNP',color_map=color_self,show=True,size=300000/adata.shape[0])
sc.pl.scatter(adata,alpha=1,x="y_array",y="x_array",color="PLP1",title='PLP1',color_map=color_self,show=True,size=300000/adata.shape[0])
```
